```python
fileNames=[]
def absoluteFilePaths(directory):
   for dirpath,_,filenames in os.walk(directory):
       for f in filenames:
           yield os.path.abspath(os.path.join(dirpath, f))
for i in range(len(sourceDir)):
   fileNames+=absoluteFilePaths(f"F:\\Dev\\MLprojects\\DeepFashion\\img\\{sourceDir[i]}")
```

```python
# this function takes in batch size and file names and returns a numpy array consisting of the pre-processed images
def dataloader(fileNames):
    return [preProcess(i) for i in fileNames]
```

```python
def preProcess(file):
    image = cv2.resize(cv2.imread(file, cv2.IMREAD_GRAYSCALE),(416,416), interpolation=cv2.INTER_CUBIC)
    label = file.split('\\')[-2]
    return (image,label)
```
